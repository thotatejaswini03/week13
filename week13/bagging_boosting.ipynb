— Boosting : it is an ensemble technique where each models are trained sequentially and each new model focus more on the mistakes made by the previous model

Why do we need boosting ?
— boosting converts many weak learners into a strong learner

Steps :
— train first weak model
— identify misclassified and highly error points
— give more importance to those misclassified points
— train next model focusing on the misclassified points
— repeat  
— combine all models into a final prediction

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import numpy as np
iris = load_iris()
x,y = iris.data, iris.target

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(max_depth=1)

# Create Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor

gbr = GradientBoostingRegressor(
    n_estimators=100,     # Number of boosting stages
    learning_rate=0.1,   # Step size
    max_depth=3,         # Depth of each tree
    random_state=42
)  
gbr.fit(x_train,y_train)
gbr_pred = gbr.predict(x_test)

# Create ADA Boosting Regressor
from sklearn.ensemble import AdaBoostRegressor

abr = AdaBoostRegressor(
    estimator=dtr,
    n_estimators=100,     # Number of boosting stages
    learning_rate=0.1,   # Step size      
    random_state=42
)  
abr.fit(x_train,y_train)
abr_pred = abr.predict(x_test)

from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error,r2_score

print("Gradient Boosting")
print("R2 score : ",r2_score(y_test,gbr_pred))
print("RMSE : ",root_mean_squared_error(y_test,gbr_pred))
print("MSE :",mean_squared_error(y_test,gbr_pred))
print("MAE :",mean_absolute_error(y_test,gbr_pred))

print("ADA Boosting")
print("R2 score : ",r2_score(y_test,abr_pred))
print("RMSE : ",root_mean_squared_error(y_test,abr_pred))
print("MSE :",mean_squared_error(y_test,abr_pred))
print("MAE :",mean_absolute_error(y_test,abr_pred))
```
