Bagging:

What is bagging?

It is an ensemble technique where multiple models train independently on diff random samples of dataset and their predictions are combined to produce final result.

Why we need bagging?

A single decision tree - have high variance, easily overfits, changes a lot if data changes lightly.Bagging reduces variance and makes predictions more stable.

How bagging works ?
Bootstrap sampling : from origin dataset create multiple new datasets, each dataset is created by random sampling with replacement , each bootstrap dataset has same size as the original
Some data points may be repeated  or left-out this is called out of bag samples.
Train multiple models : train one model for bootstrap dataset,  usually a decision tree . Each tree is trained independently.
Make Predictions : For reg : take average , for classification : take majority vote
Final Output : Aggregated prediction is more stable than a single decision tree

When we should use bagging?
When dataset is medium to large, model is overfitting, decision tree gives unstable decisions

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import fetch_california_housing
data = fetch_california_housing(as_frame=True)
df = data.frame
df.head()

x = df.drop(columns="MedHouseVal")
y = df["MedHouseVal"]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)

from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor(max_depth=4,ccp_alpha=0.01)
dtr.fit(x_train,y_train)
dtr_pred = dtr.predict(x_test)

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
print("R2 score : ",r2_score(y_test,dtr_pred))
print("mse : ",mean_squared_error(y_test,dtr_pred))
print("mae : ",mean_absolute_error(y_test,dtr_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=200,random_state=42)
rf.fit(x_train,y_train)
rf_pred = rf.predict(x_test)

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error

r2 = r2_score(y_test, rf_pred)
mse = mean_squared_error(y_test, rf_pred)
mae = mean_absolute_error(y_test, rf_pred)

print("R2 score :", r2)
print("MSE :", mse)
print("MAE :", mae)
```
